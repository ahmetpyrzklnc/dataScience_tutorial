> Veri tablolar her zaman ayn yerlerde bulunmaz. Farkl farkl alanlarda da bulunabilir. Aadaki rnekteki gibi bir tablo olabilir...



Bu tipteki verileri baz yntemlerle bir araya getirebiliriz...

```python
sozluk1 = {"sim" : ["Ahmet","Mehmet","Zeynep","Atl"], # birinci tabloyu oluturduk
          "Spor" : ["Kou","Yzme","Kou","Basketbol"],
           "Kalori" : [100,200,250,150]
          }
```

#### Merge

> Merge genel anlamda benzerlik gsteren hatta birebir ayns ama bir tane farkl stunu olan iki dataframe'i bir araya getirmeye alr. rnee bakalm:



* Burada iki farkl tablo var ve bu tablolar arasndaki tek fark aada bulunan farkl bir stun o stunu kopartp tekrardan stteki kolona nasl yazarz gelin verilerimizi ekleyelim ve adm adm gidelim

```python
mergeSozluk = {"sim" : ["Ahmet","Mehmet","Zeynep","Atl"], # birinci dataframe oluturduk...
              "Spor": ["Kou","Yzme","Kou","Basketbol"]}
```

```python
mergeDataFrame1 = pd.DataFrame(mergeSozluk)
```

```python
mergeDataFrame1
```

```python
mergeSozluk2 = {"sim" : ["Ahmet","Mehmet","Zeynep","Atl"], # ikinci dataframe oluturduk...
               "Kalori" : [100,200,250,150]}
```

```python
mergeDataFrame2 = pd.DataFrame(mergeSozluk2)
```

```python
mergeDataFrame2
```

> Yukardaki dataframeleri incelersek ortak stunlar buluruz. te biz bu stunlar tek bir dataframe olarak grmek istediimizde merge komutunu kullanrz...

```python
pd.merge(mergeDataFrame1,mergeDataFrame2, on ="sim") # birletirilmek istenen dataframeler ve iki dataframe arasnda ortak olan stun bilgisi (on)
```

```python
maasSozluk = {"sim" : ["Atl","Zeynep","Mehmet","Ahmet"],
             "Departmanlar" : ["Yazlm","Sat","Pazarlama","Yazlm"],
             "Maa" : [100,200,300,500]}
```

> Mesela bazen saysal deerler iin hesaplama yapmamz gerekebilir. Bu hesaplamalar bir fonksiyon yardmyla btn dataframede grntleyebiliriz. te bu zellii u ekilde kullanrz...

```python
maasDataFrame["Maa"].apply(maastanBrute) # burada apply() maa datasn komple maastanBrute fonksiyonu ierisine ykleyerek bir saysal analiz yapt
```

```python
maasDataFrame.isnull() # maasDataFrame'de eksik veri var mdr? yok mudur? True:var, False: yok
```

```python
yeniBirVeri = {"Karakter Snf" : ["South Park","South Park","Simpson","Simpson","Simpson"], # veri girii saladm
              "Karakter smi" : ["Cartman","Kenny","Homer","Bart","Bart"],
               "Karakter Ya" : [9,10,20,50,60]
              }
```

```python
karakterDataFrame
```

> Peki bu dataframe'i nasl pivot table olarak gsterebiliriz ?

* Pivot table'n zellii iki tane ayn deikene ait verileri toplar veya ortalamasn alr rnein aada bart'tan iki tane olduu iin iki yan ortalamasn alyor...

```python
karakterDataFrame.pivot_table(values = "Karakter Ya", index = ["Karakter Snf", "Karakter smi"]) # ortalamay alr
```

> Peki toplamasn yapamaz myz ? Evet Yapabiliriz!!

```python
karakterDataFrame.pivot_table(values = "Karakter Ya", index = ["Karakter Snf", "Karakter smi"], aggfunc = "sum") # aggfunc ile toplamay koydum
```

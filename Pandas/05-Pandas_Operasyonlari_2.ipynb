> Veri tabloları her zaman aynı yerlerde bulunmaz. Farklı farklı alanlarda da bulunabilir. Aşağıdaki örnekteki gibi bir tablo olabilir...



Bu tipteki verileri bazı yöntemlerle bir araya getirebiliriz...

```python
sozluk1 = {"İsim" : ["Ahmet","Mehmet","Zeynep","Atıl"], # birinci tabloyu oluşturduk
          "Spor" : ["Koşu","Yüzme","Koşu","Basketbol"],
           "Kalori" : [100,200,250,150]
          }
```

```python
dataFrame1 = pd.DataFrame(sozluk1, index = [0,1,2,3]) # indexi diğer tablolarda devam ettireceğim için girmeye devam ettim
```

```python
dataFrame1
```

```python
sozluk2 = {"İsim" : ["Osman","Levent","Atlas","Fatma"], # ikinci dataframe'nin verileri
          "Spor" : ["Koşu","Yüzme","Koşu","Basketbol"],
           "Kalori" : [140,210,50,180]
          }
```

```python
dataFrame2 = pd.DataFrame(sozluk2, index = [4,5,6,7])
```

```python
dataFrame2
```

```python
sozluk3 = {"İsim" : ["Ayşe","Fatma","Duygu","Nur"], # üçüncü dataframe'nin verileri
          "Spor" : ["Koşu","Yüzme","Badminton","Tenis"],
           "Kalori" : [540,190,760,480]
          }
```

```python
dataFrame3 = pd.DataFrame(sozluk3, index = [8,9,10,11])
```

#### Concatenation

> Sütun isimleri veya yapıları aynı olan yapıları alt alta yapıştırarak tek bir dataframe haline getirir.

```python
pd.concat([dataFrame1,dataFrame2,dataFrame3]) # varsayılan axis = 0'dır üst üste yazar, axis=1'se yan yana yazar
```

#### Merge

> Merge genel anlamda benzerlik gösteren hatta birebir aynısı ama bir tane farklı sütunu olan iki dataframe'i bir araya getirmeye çalışır. Örneğe bakalım:



* Burada iki farklı tablo var ve bu tablolar arasındaki tek fark aşağıda bulunan farklı bir sütun o sütunu kopartıp tekrardan üstteki kolona nasıl yazarız gelin verilerimizi ekleyelim ve adım adım gidelim

```python
mergeSozluk = {"İsim" : ["Ahmet","Mehmet","Zeynep","Atıl"], # birinci dataframe oluşturduk...
              "Spor": ["Koşu","Yüzme","Koşu","Basketbol"]}
```

```python
mergeDataFrame1 = pd.DataFrame(mergeSozluk)
```

```python
mergeDataFrame1
```

```python
mergeSozluk2 = {"İsim" : ["Ahmet","Mehmet","Zeynep","Atıl"], # ikinci dataframe oluşturduk...
               "Kalori" : [100,200,250,150]}
```

```python
mergeDataFrame2 = pd.DataFrame(mergeSozluk2)
```

```python
mergeDataFrame2
```

> Yukarıdaki dataframeleri incelersek ortak sütunlar buluruz. İşte biz bu sütunları tek bir dataframe olarak görmek istediğimizde merge komutunu kullanırız...

```python
pd.merge(mergeDataFrame1,mergeDataFrame2, on ="İsim") # birleştirilmek istenen dataframeler ve iki dataframe arasında ortak olan sütun bilgisi (on)
```

```python
maasSozluk = {"İsim" : ["Atıl","Zeynep","Mehmet","Ahmet"],
             "Departmanlar" : ["Yazılım","Satış","Pazarlama","Yazılım"],
             "Maaş" : [100,200,300,500]}
```

```python
maasDataFrame = pd.DataFrame(maasSozluk)
```

```python
maasDataFrame
```

> Bu veri setinde analiz yapmak kolay ama atıyorum 5000 satırlı bir veri setiyle çalışıyorsanız ve Departmanlardan kaç tane olduğunu görmek istediğinizde bunu görebilmek oldukça zordur. Bu yüzden dolayı bu tip sorunları şu şekilde çözümleyebiliriz...

```python
maasDataFrame["Departmanlar"].unique() # her bir veriden sadece bir tanesini getirir....
```

```python
maasDataFrame["Departmanlar"].nunique() # her bir departmanların sayısını verecektir  toplamda 4 veri var ama yazılım tekrar ettiği için bir tanesini
# dahil etti
```

```python
maasDataFrame["Departmanlar"].value_counts() # hangi departmandan kaç kere tekrar etti sorusunu çözümler...
```

> Mesela bazen sayısal değerler için hesaplama yapmamız gerekebilir. Bu hesaplamaları bir fonksiyon yardımıyla bütün dataframede görüntüleyebiliriz. İşte bu özelliği şu şekilde kullanırız...

```python
def maastanBrute(maas):
    return maas * 0.66
```

```python
maasDataFrame["Maaş"].apply(maastanBrute) # burada apply() maaş datasını komple maastanBrute fonksiyonu içerisine yükleyerek bir sayısal analiz yaptı
```

```python
maasDataFrame.isnull() # maasDataFrame'de eksik veri var mıdır? yok mudur? True:var, False: yok
```

```python
yeniBirVeri = {"Karakter Sınıfı" : ["South Park","South Park","Simpson","Simpson","Simpson"], # veri girişi sağladım
              "Karakter İsmi" : ["Cartman","Kenny","Homer","Bart","Bart"],
               "Karakter Yaş" : [9,10,20,50,60]
              }
```

```python
karakterDataFrame = pd.DataFrame(yeniBirVeri)
```

```python
karakterDataFrame
```

> Peki bu dataframe'i nasıl pivot table olarak gösterebiliriz ?

* Pivot table'ın özelliği iki tane aynı değişkene ait verileri toplar veya ortalamasını alır Örneğin aşağıda bart'tan iki tane olduğu için iki yaşın ortalamasını alıyor...

```python
karakterDataFrame.pivot_table(values = "Karakter Yaş", index = ["Karakter Sınıfı", "Karakter İsmi"]) # ortalamayı alır
```

> Peki toplamasını yapamaz mıyız ? Evet Yapabiliriz!!

```python
karakterDataFrame.pivot_table(values = "Karakter Yaş", index = ["Karakter Sınıfı", "Karakter İsmi"], aggfunc = "sum") # aggfunc ile toplamayı koydum
```

```python
y = df["Type"].values
x = df.drop("Type", axis = 1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping # overfitting problemlerini çözmek için kullanılır
```

```python
x_train.shape
```

* Genel geçer kurala göre bir model kurulacağı zaman giriş layerına veri setinde kaç sütun varsa o kadar nöron oluşturulması tavsiye edilir ve diğer ara katman layerlarına ise çıkış katmanıyla giriş katmanı arasında bir nöron sayısı koyulması tavsiye edilir.
* Ayrıca sınıflandırma problemlerinde çıkış katmanında ise bir aktivasyon tanımlanabilir. Bu aktivasyon ise en iyisi sigmoid'tir. Sigmouid bize 0,1 arasında değerler döndürür.

```python
model = Sequential()

model.add(Dense(units = 30, activation="relu"))
model.add(Dense(units = 30, activation="relu"))
model.add(Dense(units=30, activation="relu"))

model.add(Dense(units=1, activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer = "adam") # sınıflandırma için binary_crossentropy kullandık
```

> Aşırı öğrenmeye örnek olması açısından epochs değerini abartacağım

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data = (x_test,y_test), verbose = 1) 
```

```python
model.history.history
```

```python
veri_kaybi = pd.DataFrame(model.history.history)
```

```python
veri_kaybi.plot()
```

```python
model = Sequential()

model.add(Dense(units = 30, activation="relu"))
model.add(Dense(units = 30, activation="relu"))
model.add(Dense(units=30, activation="relu"))

model.add(Dense(units=1, activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer = "adam") 
```

```python
earlyStopping = EarlyStopping(monitor="val_loss", mode="min", verbose=1, patience=25) # val_lossu takip et min değerde dur
```

```python
model.fit(x=x_train, y=y_train, epochs = 700, validation_data=(x_test,y_test), verbose = 1, callbacks=[earlyStopping])
```

```python
veri_kaybi = pd.DataFrame(model.history.history)
```

```python
veri_kaybi.plot()
```

```python
model = Sequential()

model.add(Dense(units = 30, activation="relu")) 
model.add(Dropout(0.6)) # modelin doğruluk sonucuna göre overffitting olmaya başladığı anda katmandan çıkacak son katman hariç her katman arasına konur

model.add(Dense(units = 30, activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=30, activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=1, activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer = "adam") 
```

```python
model.fit(x=x_train, y=y_train, epochs = 700, validation_data=(x_test,y_test), verbose = 1, callbacks=[earlyStopping])
```

```python
kayipDf = pd.DataFrame(model.history.history)
```

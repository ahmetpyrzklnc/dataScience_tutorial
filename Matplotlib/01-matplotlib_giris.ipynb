> plot genel anlamda bize bir grafik verir. Ve bu grafik üç adet parametre alır bu parametreler x ekseni y ekseni ve çizginin rengi şeklindedir.

* Aşağıdaki örneği inceleyelim
* Aşağıdaki örnekte plot özelliği üç parametre aldı birinci parametrede girilen numpy dizisini grafikte x eksenine koydu (numpyYasListesi)
* ikinci parametre girilen numpy dizisini (numpyKiloListesi)'ni aldı y parametresine koydu.
* Son girilen parametre bize grafik çizgisinin rengini belirtir.

```python
plt.plot(numpyYasListesi,numpyKiloListesi, "g") # r: red, g: green, b: blue ---> istediğin rengin baş harfini yazarak rengini ayarlar
```

> peki eksenlere isimleri nasıl verebiliriz ? 

```python
plt.plot(numpyYasListesi,numpyKiloListesi, "r")
plt.xlabel("Yaş") # x ekseni
plt.ylabel("Kilo") # y ekseni
#plt.show() # yukarıda %matplotlib inline yazdığımız için yazmıyoruz normalde yazılması gerekir
```

> Peki çıkan grafiğe başlık atayabilir miyiz?

```python
plt.plot(numpyYasListesi,numpyKiloListesi, "b")
plt.xlabel("Yaş") # x ekseni
plt.ylabel("Kilo") # y ekseni
plt.title("Kilonun Yaşa Göre Değişim Grafiği") # başlık 
```

```python
numpyDizisi1 = np.linspace(0,10,20) # 0-10 arasında 20 adet veri üret
```

```python
numpyDizisi1 # eşit aralıklarla 20 adet sayı üretir
```

```python
numpyDizisi2 = numpyDizisi1 ** 3 # numpy dizisinin küpünü alır
```

```python
numpyDizisi2
```

> Grafiğini gösterelim 

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g")# konveks bir grafik çizdirir
plt.title("Konveks Yapıda Grafik")
```

```python
plt.plot(numpyDizisi2,numpyDizisi1,"r") # konkav bir grafik çizdirir
plt.title("Konkav Yapıda Grafik")
```

* g*- : green, yıldız noktalı bir grafik oluşturur.
* g-- : kesik çizgili bir grafik oluşturur..

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g*-") # yıldızların her biri veri noktasını işaret eder
plt.title("Yıldız noktalı konveks grafik")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
plt.title("Kesik çizgili konveks bir grafik")
```

> Aynı anda birden fazla grafiği yan yana çizmemiz de matplotlib ile mümkündür. Bunun için subplot() özelliğini kullanırız.

Adım adım gidecek olursak: 
* subplot(kaç tane satır olacak, kaç tane kolon olacak, kaçıncı grafik) şeklinde bir yapı oluşturulur.

```python
plt.subplot(1,2,1) # bir satır olacak, 2 kolon olacak, birinci grafik
plt.plot(numpyDizisi1,numpyDizisi2, "r*-")
plt.title("Konveks Grafik")
plt.subplot(1,2,2)
plt.plot(numpyDizisi2,numpyDizisi1, "g--") # bir satır olacak, 2 kolon olacak, ikinci grafik
plt.title("Konkav Grafik")
```

> Bu zamana kadar sürekli varsayılan plt grafiklerini çizdirmiştik.
Bu grafikleri kendimize göre özelleştirebiliriz. Neyden bahsediyorum ?
Örneğin:
* Grafiklerin boyutlarını kendim ayarlamak isteyebilirim 
* Grafiklerin x ve y boyutlarını kendim ayarlayabilirim
İşte bütün bu özellikleri **figure()** özelliğini kullanarak gerçekleştirebilirim...

```python
benimFigure = plt.figure() # figure nesnesini oluşturdum.
benimFigureAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9]) # son iki tanesi grafiğin boyutlarıdır. width, height
benimFigureAxes.plot(numpyDizisi1,numpyDizisi2, "g")
benimFigureAxes.set_xlabel("X ekseni Veri İsmi")
benimFigureAxes.set_ylabel("Y ekseni Veri İsmi")
benimFigureAxes.set_title("Grafik Başlığı")
```

#### İç içe grafikler

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.2,0.6,0.3,0.3]) # x ekseni başlangıç - y ekseni başlangıç --> iç içe grafiklerde konumun yerini oynayabilirsiniz 0.2,0.6

eksen1.plot(numpyDizisi1,numpyDizisi2, "g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana Grafik Başlığı")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X ekseni küçük")
eksen2.set_ylabel("Y ekseni küçük")
eksen2.set_title("Alt Grafik Başlığı")
```

> Bu işlemlerin aynısını subplotslarla da gerçekleştirebiliriz....

```python
plt.subplots() # axes ve figure döndürür --> tuple döndürür
```

```python
plt.figure() # sadece figure döndürür axes döndürmez 
```

* **figure()** ile **subplots()** en temel fark ikisi arasında biri hem axes hemde figure döndürürken (**subplots()**), diğeri ise sadece **figure()** döndürür... 

* Buna göre o zaman bir subplots içerisinde hem ekseni hemde figure oluşturmayı deneyelim

```python
(benimFigure, benimEksen) = plt.subplots() # subplots içerisinde hem figure oluşturdum hemde axes yani eksenler oluşturdum
```

```python
benimFigure # figure'ü çağırdım..
```

```python
benimEksen # eksenleri yani axesleri çağırdım burada aynı zamanda axes değerleri verebilirim...
```

```python
(benimFigure, benimEksen) = plt.subplots(nrows=1,ncols=2) # parametre içerisine değer verdim
# benimEksen.plot(numpyDizisi1,numpyDizisi2, "r") --> arraye dönüştüğü için for loopla işlemen gerekir
type(benimEksen)
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1,ncols=2) 

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2, "g")
    eksen.set_xlabel("X ekseni")
plt.tight_layout() # grafikler arasındaki mesafeyi açar
```

> İki farklı veriden oluşan grafik çizdirmek istersek

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1, ncols=2) 

for i, eksen in enumerate(benimEksenler):
    if i == 0:  
        eksen.plot(numpyDizisi1, numpyDizisi2, "g")
        eksen.set_title("Birinci Grafik")
    elif i == 1:  
        eksen.plot(numpyDizisi2, numpyDizisi1, "r")
        eksen.set_title("İkinci Grafik")
    eksen.set_xlabel("X ekseni")

plt.tight_layout()
```

#### legend()

```python
yeniFigur = plt.figure() # nesneyi tanımladım

yeni_eksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeni_eksen.plot(numpyDizisi1,numpyDizisi2, label = "numpyDizisi ** 2") #label kullanarak çizgiye bir ad verdik
yeni_eksen.legend() # eğrinin ne anlama geldiğini grafik üzerinde göstermeye çalışır
```

> Bunun amacı aynı grafik üzerinde birden fazla eğri varsa bu eğrilerin ne anlama geldiğini bize gösterir

```python
yeniFigur = plt.figure() # dpi=300 ise kaliteyi daha da güzelleştirir figsize=(3,3) --> grafiğin boyutlarını bu şekilde de girebilirsin

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1, numpyDizisi2 ** 2, label = "numpydizisi 2 üstü karesi eğrisi")
yeniEksen.plot(numpyDizisi1, numpyDizisi2 ** 3, label = "numpydizisi1 2in küpü eğrisi")
yeniEksen.legend(loc=1) # lokasyonu istediğin gibi 1,2,3,4,5,6 diye girebilirsin bu kod bilgi kutusunun yerini değiştirir...
```

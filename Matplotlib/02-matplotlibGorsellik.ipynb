```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
numpyDizisi2
```

Renk kodlarını kendi grafiğine göre özelleştirebilirsin. Bunun için rgb hex code'nu **"color=#fff"** diye belirtmen gerekir.
* Bu belirteçle birlikte çizgilerin renklerini istediğin gibi girebilirsin
* Grafik çizgilerinin olduğu renklere saydamlık katmak istiyorsan **"alpha=0.3"** şeklinde girebilirsin
* Grafik çizgilerinin kalınlığını belirtmek istiyorsan "**linewidth=0.4**" şeklinde değer girebilirsin.

```python
(benimFigur, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1, numpyDizisi2, color="#3A95A8") # color hex code rengine göre çizgi rengini değiştirdi...
benimEksen.plot(numpyDizisi2, numpyDizisi1, color = "#C96F23", alpha=0.3) # alpha saydamlık kattı varsayılan değer 1'dir
```

> **"linewidth=0.5"** grafiğine bi yapmaya çalışalım

```python
(YeniFigur, yeniEksen) = plt.subplots()

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 3, "blue" , linewidth = 6.0) # kalın bir çizgi
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 2, "yellow", linewidth = 0.5)
```

> **"linestyle"** grafik çizgilerini değiştirmemize olanak sağlar

```python
(YeniFigur, yeniEksen) = plt.subplots()

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 3, "blue" , linewidth = 2.0) 
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 2, "yellow", linewidth = 3.0)

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 4, color = "#C96F23", linestyle = "--") # çizgi çizgi stile sahip bir çizgi çizer
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 5, color = "#C96F23", linestyle = ":") # nokta nokta stile sahip bir çizgi çizdi...
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 5, color = "#C96F23", linestyle = "-.") # çizgi nokta stile sahip bir çizgi çizdi...
```

* **"marker"** grafik çizgilerinde veri noktalarını işaretlemeye olanak sağlar.

* **"markersize"** grafik çizgisindeki markerların boyutunu ayarlayabiliriz.

* **"markerfacecolor"** grafik çizgisindeki markerların rengini ayarlar.

```python
(YeniFigur, yeniEksen) = plt.subplots()

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 3, "blue" , linewidth = 2.0) 
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 2, "yellow", linewidth = 3.0)

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 4, color = "#C96F23", linestyle = "--") # çizgi çizgi stile sahip bir çizgi çizer
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 5, color = "#C96F23", linestyle = ":") # nokta nokta stile sahip bir çizgi çizdi...
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 6, color = "#C96F23", linestyle = "-.") # çizgi nokta stile sahip bir çizgi çizdi...

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 7, color = "#000000", linestyle = "--", marker = "+") # veri noktaları üzerinde + işareti koydu
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 8, color = "#000000", linestyle = "--", marker = "o", markersize = 4, markerfacecolor="red") # marker kırmızı
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 8, color = "#000000", linestyle = "--", marker = "*", markersize = 8, markerfacecolor="blue") #marker boyut
```
